```python
import matplotlib.pyplot as plt
import cv2 


image = cv2.imread("IllusionDiffusion_output/orchid.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.title("Orchid flower generated by Illusion Diffusion")
plt.axis("off")
plt.imshow(image)
plt.show()
```

```python
import requests

dotenv = dotenv.load_dotenv()
PLANT_API_KEY = os.getenv("PLANT_API_KEY")

url = "https://my-api.plantnet.org/v2/identify/all"

headers = {}
payload = {
    "api-key" : PLANT_API_KEY
    }

files = {
    "images" : open("IllusionDiffusion_output/orchid.jpg", "rb")
}
try:
    response = requests.post(url, headers = headers, params = payload, files = files)
    response.raise_for_status
except requests.exceptions.HTTPError as errh:
    print(f"HTTP Error occurred: {errh}")     
except requests.exceptions.RequestException as err:
    print(f"An error occurred: {err}")        
else:    
    response.status_code
```

```python
flower_name = response.json().get('results')[0]
flower_name = flower_name['species']['commonNames']
flower_name
```
